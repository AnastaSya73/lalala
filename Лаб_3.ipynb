```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/student-por.csv')
df
```

```python
def create_holes(dataframe, column_name, hole_percentage = 10):
    if hole_percentage < 0 or hole_percentage > 100:
        raise ValueError("Процент дырявости должен быть в диапазоне от 0 до 100.")

    dataframe_with_holes = dataframe.copy()
    num_holes = int(len(dataframe) * (hole_percentage / 100))
    hole_indices = np.random.choice(len(dataframe), num_holes, replace=False)
    dataframe_with_holes.loc[hole_indices, column_name] = np.nan

    return dataframe_with_holes

df1 = create_holes(df, 'age')# Добавление пропущенных значений, т.к. в исходном датасете они отсутствуют
df2 = create_holes(df1, 'school')
df2
```

```python
df_cleaned = df2.dropna()#Удаление
df_cleaned
```

```python
df_filled_median = df2.fillna(df2.median())

df_filled_median #Заполнение при помощи медианы
```

```python
df_filled_mean = df2.fillna(df2.mean())

df_filled_mean#Заполнение при помощи среднего
```

```python
df_filled_mode = df2.fillna(df2["school"].mode()[0])

df_filled_mode#Заполнение при помощи моды
```

```python
df_interpolated = df2.interpolate()

df_interpolated #Заполнение при помощи интерполяции
```

##Поиск выбросов и аномалий

**Те, кто выполняет задания со * делают поиск выбросов с помощью написанного ранее класса**

Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


```python
mean = df["age"].mean()
std_deviation = df["age"].std()
threshold = 3 * std_deviation

# Идентифицируем выбросы
outliers = [x for x in df["age"] if abs(x - mean) > threshold]

# Выводим результаты
print("Среднее значение данных:", mean)
print("Стандартное отклонение данных:", std_deviation)
print("Потенциальные выбросы:", outliers) #Методом среднего
```

ОБЪЯСНЕНИЕ: этот метод нашел следующие выбросы: возраст = 22 и 21, это логично ведь людей данного возраста меньше всего в датасете, хоть и говорят что учиться никогда не поздно, но людей которые идут на этот отчаянный шаг не так уж и много.

```python
Q1 = df["age"].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers2 = [x1 for x1 in df["age"] if ((x1 < lower) or (x1 > upper))]#При помощи межквартильного размаха

print("Потенциальные выбросы:", outliers2)
```

ОБЪЯСНЕНИЕ: этот метод нашел следующие выбросы: возраст = 22, это логично ведь людей данного возраста очень мало в датасете, хоть и говорят что учиться никогда не поздно, но людей которые идут на этот отчаянный шаг не так уж и много.

ВЫВОД: после поиска выбросов двумя разными методами, был получен следующий результат: метод основанный на среднем, посчитал что выбросами являются все люди достигшие 21 и 22 летного возраста, а метод основанный на межквартильном размахе выбросами посчитал только людей в возрасте 22 лет.

```python
df.age.describe()
```

```python
df3=df.copy()

upper_array = np.where(df3['age']>upper)[0]

df3.drop(index=upper_array, inplace=True)#Удаление выбросов по методу межквартильного размаха

df3
```

```python
df3.age.describe()
```

```python
df4=df.copy()

upper_array1 = np.where(df4['age'] - mean > threshold)[0]

df4.drop(index=upper_array1, inplace=True)#Удаление выбросов по методу среднего

df4
```

```python
df4.age.describe()#Статистика и объяснение
```

**Объяснения**

Таким образом, мы видим, что с поиском выбросов лучше справился метод межквартильного размаха, так как после удаления с его помощью выбросов, статистика изменилась не так сильно, как после удаления выбросов по методу среднего.

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```

```python
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}
df = pd.DataFrame(data)
df
```

```python
corr_pearson(df["Рост"], df["Вес"])
```
